# Chapter 2

```python
state_union = nltk.corpus.state_union
state_union
```

```python
men_women_ppl = ['men', 'women' , 'people']
```

```python
men_women_ppl_fdist = nltk.FreqDist(word.lower() for word in state_union.words())
for modal in men_women_ppl:
    print ' '.join([modal, str(men_women_ppl_fdist[modal])])+"\n"
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(15,5))
cfd = nltk.ConditionalFreqDist(
    (target, fileid[:4])
    for fileid in state_union.fileids()
    for w in state_union.words(fileid)
    for target in ['men', 'women', 'people']
    if w.lower().startswith(target))
cfd.plot()
```

```python
from __future__ import division  # Python 2 users only
import nltk, re, pprint
from nltk import word_tokenize
```

```python
path = nltk.data.find("/Users/ravirajmulasa/Documents/Certificate_in_MachineLearning/Advanced_Machine_Learning/Homework/HW4/online_books.txt")
file_content = open(path).read().decode('string_escape').decode("utf-8")
```

```python
monetary_pattern = r'''(?x)USD[\d,]+(.\d\d)?|EUR[\d.]+(,\d\d)?|INR[\d.]+(,\d\d)?|JPY[\d]+'''
nltk.regexp_tokenize(file_content, monetary_pattern)
```

```python
number_pattern = r'''(?x)\d+,\d+'''
nltk.regexp_tokenize(file_content, number_pattern)
```

```python
name_pattern = r'''(?x)[A-Z][a-z]+\s[A-Z][a-z]+'''
nltk.regexp_tokenize(file_content, name_pattern)
```

```python
date_pattern = r'''(?x)\d{2}.\d{2}.\d{4}|\d{4}/\d{2}/\d{2}|\d.\syears|\d.\sdays|\d.\smonths'''
nltk.regexp_tokenize(file_content, date_pattern)
```
